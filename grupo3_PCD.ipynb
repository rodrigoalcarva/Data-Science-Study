## Dataset

O DataSet que iremos analisar é o  **`PetFinder_dataset.csv`**.
Para complementar a informação, teremos mais três DataSets: **`Breed_labels.csv`**, **`Colors_labels.csv`**, **`States_labels.csv`**.

Este DataSet está disponível para fazer o download no [Kaggle](https://www.kaggle.com), e contém dados do [PetFinder.my Adoption Prediction](https://www.kaggle.com/c/petfinder-adoption-prediction/overview). 


#### File descriptions:
    
- PetFinder_dataset.csv - Dados Tabulares/textuais 
- breed_labels.csv - Contém Type e BreedName para cada BreedID. Type 1 corresponde a um cão e o Type 2 corresponde a um gato.
- color_labels.csv - Contém a ColorName para cada ColorID.
- state_labels.csv - Contém o StateName para cada StateID.


#### Target:
AdoptionSpeed - Rapidez de Adoção: quanto menor mais rápida foi a adoção.

O valor é determinado pela rapidez com que o animal é adotado. Os valores podem ser:
- 0 - O animal foi adotado no dia em que foi registado.
- 1 - O animal foi adotado 1 a 7 dias (1ª Semana) após ser registado.
- 2 - O animal foi adotado 8 a 30 dias (1º Mês) após ser registado.
- 3 - O animal foi adotado 31 a 90 dias (1º e 3º Mês) após ser registado.
- 4 - O animal foi adotado 100 dias após ser registado. (Não há animais, neste dataset, que tenham sido adotados entre os 90 e os 100 dias após o seu registo).

#### Data Fields: 
- PetID - Identificador único de cada animal
- Type - Tipo de Animal (1 = Cão, 2 = Gato)
- Name - Nome do Animal (Vazio se não tiver nome)
- Age - Idade, em meses, de quando o animal foi registado
- Breed1 - Raça principal do animal (Referente ao dicionário BreedLabels)
- Breed2 - Raça secundária do animal, caso este seja cruzado (Referente ao dicionário BreedLabels)
- Gender - Género do Animal (1 = Macho, 2 = Fêmea, 3 = Ninhada)
- Color1 - Cor 1 do animal (Referent ao dicionário ColorLabels)
- Color2 - Cor 2 do animal (Referent ao dicionário ColorLabels)
- Color3 - Cor 3 do animal (Referent ao dicionário ColorLabels)
- MaturitySize - Tamanho na Maturidade (1 = Pequeno, 2 = Médio, 3 = Grande, 4 = Muito Grande, 0 = Não especificado)
- FurLength - Tamanho do Pêlo (1 = Pequeno, 2 = Médio, 3 = Longo, 0 = Não especificado)
- Vaccinated - O animal foi vacinado (1 = Sim, 2 = Não, 3 = Não Sabe)
- Dewormed - O animal foi desparasitado (1 = Sim, 2 = Não, 3 = Não Sabe)
- Sterilized - O animal foi castrado (1 = Sim, 2 = Não, 3 = Não Sabe)
- Health - Condição de Saúde (1 = Saudável, 2 = Pequena Lesão, 3 = Lesão Grave, 0 = Não Especificado)
- Quantity - Número de animais representados no perfil
- Fee - Taxa de Adoção (0 = Gratuito)
- State - Localização do Estado na Malásia (Referente ao dicionário StateLabels)
- RescuerID - Identificador único de quem resgatou o animal
- VideoAmt - Total de vídeos do animal
- PhotoAmt - Total de fotos do animal
- Description - Descrição do perfil do animal.

```python
main_count = df['AdoptionSpeed'].value_counts(normalize=True).sort_index()
def prepare_plot_dict(df, col, main_count):
    """
    Esta função serve para preparar o dicionário 
    com a data para o plotting.
    
    Com isto, queremos mostrar como as diferentes variáveis influenciam
    a rapidez da Adoção. 
    At first I calculate base rates, then for each category in the 
    column I calculate rates of Adoption speed and find difference 
    with the base rates.
    
    """
    main_count = dict(main_count)
    plot_dict = {}
    for i in df[col].unique():
        val_count = dict(df.loc[df[col] == i, 'AdoptionSpeed'].value_counts().sort_index())

        for k, v in main_count.items():
            if k in val_count:
                plot_dict[val_count[k]] = ((val_count[k] / sum(val_count.values())) / main_count[k]) * 100 - 100
            else:
                plot_dict[0] = 0

    return plot_dict

def make_count_plot(df, x, hue='AdoptionSpeed', title='', main_count=main_count):
    """
    Plotting countplot with correct annotations.
    """
    g = sb.countplot(x=x, data=df, hue=hue);
    plt.title(f'AdoptionSpeed {title}');
    ax = g.axes

    plot_dict = prepare_plot_dict(df, x, main_count)

    for p in ax.patches:
        h = p.get_height() if str(p.get_height()) != 'nan' else 0
        text = f"{plot_dict[h]:.0f}%" if plot_dict[h] < 0 else f"+{plot_dict[h]:.0f}%"
        ax.annotate(text, (p.get_x() + p.get_width() / 2., h),
             ha='center', va='center', fontsize=11, color='green' if plot_dict[h] > 0 else 'red', rotation=0, xytext=(0, 10),
             textcoords='offset points')  

plt.figure(figsize=(18, 8));
make_count_plot(df=df, x='Type', title='by pet Type')
```

Após a análise dos dois gráficos acima apresentados, verificamos que, embora existam mais cães registados para adoção, os gatos são mais prováveis de ser adotados.

```python
def make_factor_plot(df, x, col, title, main_count=main_count, hue=None, ann=True, col_wrap=4):
    """
    Plotting countplot.
    Making annotations is a bit more complicated, 
    because we need to iterate over axes.
    """
    if hue:
        g = sb.factorplot(col, col=x, data=df, kind='count', col_wrap=col_wrap, hue=hue);
    else:
        g = sb.factorplot(col, col=x, data=df, kind='count', col_wrap=col_wrap);
    plt.subplots_adjust(top=0.9);
    plt.suptitle(title);
    ax = g.axes
    plot_dict = prepare_plot_dict(df, x, main_count)
    if ann:
        for a in ax:
            for p in a.patches:
                text = f"{plot_dict[p.get_height()]:.0f}%" if plot_dict[p.get_height()] < 0 else f"+{plot_dict[p.get_height()]:.0f}%"
                a.annotate(text, (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=11, color='green' if plot_dict[p.get_height()] > 0 else 'red', rotation=0, xytext=(0, 10),
                     textcoords='offset points')  
```

## 3.1 Linear Regression With statsmodels

Uma vez que, neste trasbalho são desejados grandes níveis de detalhe decidimos recorrer ao uso da Regressão Linear com statsmodel. Mais concretamente com OLS. OLS significa Ordinary Least Squares. Este é um modelo muito usado em econometria pois possui um conceito matemático simples, que facilita a explicação para não-técnicos, como é o nosso caso.

Para fazermos melhor esta análise recorremos também a uma medida de ajuste para modelos de regressão linear, o coeficiente de determinação. Esta medida varia entre 0 e 1 e, quanto maiores forem os valores, maior será a precisão do modelo.

```python
import statsmodels.api as sm
X = df.iloc[:,0:14]
Y = df.loc[:,"AdoptionSpeed"]
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
print('coefficient of determination:', model.rsquared)
print('adjusted coefficient of determination:', model.rsquared_adj)
print('regression coefficients:', model.params)
print('predicted response:', model.fittedvalues, sep='\n')
print('predicted response:', model.predict(X), sep='\n')
```

Após a análise dos dados obtidos, verificamos que os valores previstos se encontram bastante semelhantes aos valores obtidos. Porém, o coeficiente de determinação tem um valor de 0,062, o que é um valor bastante baixo, ou seja, o modelo não é muito preciso. 
Assim, avaliando toda esta informação decidimos não continuar com o estudo do nosso dataset através deste tipo de regressão.

## 3.5 Cluster Analysis

Uma vez que o nosso método anteriror, alternativo às Regressões Lineares, falhou decidimos tentar usar a Análise de Clusters para corroborar algumas das nossas conclusões.

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, AffinityPropagation #For clustering
from sklearn.mixture import GaussianMixture
#K means Clustering 
def doKmeans(X, nclust=5):
    model = KMeans(nclust)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = doKmeans(df, 5)
kmeans = pd.DataFrame(clust_labels)
df.insert((df.shape[1]),'kmeans',kmeans)
```

```python
#Plot the clusters obtained using k means
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(df['Fee'],df['Age'],
                     c=kmeans[0],s=50)
ax.set_title('K-Means Clustering')
ax.set_xlabel('Fee')
ax.set_ylabel('Age')
plt.colorbar(scatter)
```

Com o intuito de analisar como o feature Fee e o feature Age influenciam a Adoption Speed colocámos ambos nesta análise de clusters. Temos 5 clusters, cada um correspondente a um dos períodos de adoção.
Ao observarmos este gráfico verificamos que, as nossas previsões anteriores (as da secção 2.15) estavam corretas: As pessoas tendem a adotar animais mais novos e com taxas abaixo dos $100 muito mais rapidamente.
